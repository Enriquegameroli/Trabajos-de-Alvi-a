#Actividad 9 - Traductor
Traduce palabras de ESAÑOL A INGLÉS usando un diccionario de Python

```python
dic = {
    "hola": "hello",
    "mundo": "world",
    "casa": "house",
    "perro": "dog",
    "gato": "cat",
    "escuela": "school",
    "libro": "book",
    "agua": "water",
    "comida": "food",
    "amigo": "friend"
}
```

```python
# definir la función para traducir
def translate(word):
    return dic.get(word, "no encontrado")
```

```python
# Prueba
world = "casa"
print(translate(world))
```
